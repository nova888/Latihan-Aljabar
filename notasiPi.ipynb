```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definisikan Pi
pi = np.pi

# 2. Buat rentang x dari -4*pi sampai 4*pi dengan 400 titik agar halus
x = np.linspace(-4*pi, 4*pi, 400)

# 3. Hitung y (sin(x)/x)
# Kita gunakan np.sinc(x/pi) dari numpy atau manual seperti di bawah:
y = np.sin(x) / x

# 4. Plot grafik
plt.figure(figsize=(10, 5))
plt.plot(x, y, label=r'$f(x) = \frac{\sin(x)}{x}$')

# Tambahkan garis bantu dan label
plt.axhline(0, color='black', lw=1) # Garis horizontal y=0
plt.axvline(0, color='black', lw=1) # Garis vertikal x=0
plt.title("Grafik Fungsi Sinc")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

# 5. Tampilkan
plt.show()
```
